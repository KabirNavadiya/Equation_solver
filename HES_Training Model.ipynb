```python
from google.colab import drive
import cv2
import numpy as np


def load_images_from_folder(file_paths):
    train_data = []
    for file_path in file_paths:
        img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
        img = ~img
        if img is not None:
            ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

            ctrs, ret = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
            cnt = sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[0])
            w, h = int(28), int(28)
            maxi = 0
            for c in cnt:
                x, y, w, h = cv2.boundingRect(c)
                maxi = max(w * h, maxi)
                if maxi == w * h:
                    x_max, y_max, w_max, h_max = x, y, w, h
            im_crop = thresh[y_max:y_max + h_max + 10, x_max:x_max + w_max + 10]
            im_resize = cv2.resize(im_crop, (28, 28))
            im_resize = np.reshape(im_resize, (784, 1))
            train_data.append(im_resize)
    return train_data

```

```python
import pickle

import keras
from keras.models import Model
from keras.layers import *
from keras import optimizers
from keras.layers import Input, Dense
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from tensorflow.keras.utils import to_categorical

from tensorflow.keras import backend as K

from keras.models import model_from_json
```

Reshape l to have the correct shape

```python
np.random.seed(1212)
K.set_image_data_format('channels_last')
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/IPCV/dataset.csv', index_col = False)
labels = df_train[['784']]
```

```python
df_train.drop(df_train.columns[[784]], axis = 1, inplace = True)

df_train.head()
print(len(df_train))
```

```python
# change classes if add any new images to train model
# num_classes = total number of different image classes to train model
y_train = to_categorical(labels, num_classes = 19)
```

```python
# Reshape input data
x_train = np.array([np.array(df_train.iloc[i]).reshape(28, 28, 1) for i in range(len(df_train))])
```

```python
print(x_train[0].shape)
print(y_train[0].shape)
```

```python


# Create and compile the model
model = Sequential()
model.add(Conv2D(30, (5, 5), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(15, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(19, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# history = model.fit(l, cat, epochs = 15, batch_size = 200, shuffle = True, verbose = 1)

# Train the model
history = model.fit(x_train, y_train, epochs=15, batch_size=200, shuffle=True, verbose=1)
```

```python
model_json = model.to_json()

with open("model.json", "w") as json_file:
    json_file.write(model_json)

#Serializing weights to HDF5
model.save_weights("model.h5")
```
